```python
data_raw=pandas.read_excel(fname)
data_raw.to_latex("raw_data.tex",index=False)
data_raw
```

```python
data=data_raw.copy()[["Total Population","Infected cases","Black/African American"]]
data.index=["R1","R2"]
data.rename(columns={"Total Population":"Population","Black/African American":"D1","Infected cases":"Infected"},inplace=True)
data["D2"]=data["Population"]-data["D1"]
data.index.name="Regions"
data.to_latex("reduced_data.tex")
data
```

```python
class analysis:
    data=data
    demographics_colnames=["D1","D2"]

    
    def __init__(self,d_infected,d_population):
        self.d_infected=d_infected
        self.d_population=d_population
        self.demographics=data.loc[:,self.demographics_colnames].sum(axis="index")
        self.demographics/=self.demographics.sum()
        
    def compute_demand(self,normalize=False):
        out=self.d_infected*self.data["Infected"]+self.d_population*self.data["Population"]
        if normalize:
            out/=out.sum()
        out.name="demand"
        return out
    
    def compute_impact(self,include_gap=False):
        demand=self.compute_demand(normalize=True)
        impact=0
        for r in data.index:
            impact+=self.data.loc[r,self.demographics_colnames]*demand[r]
        impact/=impact.sum()
        out=impact
        if include_gap:
            diff=self.demographics-impact
            out=(out,numpy.linalg.norm(diff.values,ord=1))
        return out
    
    def compute_health_gap(self,t_1,t_2):
        demand=self.compute_demand()
        temp=numpy.array([t_1/demand["R1"],t_2/demand["R2"]])
        R=numpy.mean(temp)
        shortfall=numpy.mean(numpy.maximum(R-temp,0))/R
        return shortfall
        
        
        

(d_infected,d_population)=(0.1,0.9)
a=analysis(d_infected,d_population)
justice=(a.demographics["D1"],a.demographics["D2"])
print("demand=",a.compute_demand(normalize=True))
print("impact=\n",a.compute_impact(include_gap=True))
```

```python
plt.figure()
plt.plot([0,1],[1,0],color="blue")
plt.xlabel("Demographic 1")
plt.ylabel("Demographic 2")
plt.scatter(justice[0],justice[1],color="green",s=200,label=r"$J^*$")
title=[]
title.append("Social Justice")
plt.title("\n".join(title))
plt.legend()
saver("socialjustice")
plt.show()
plt.close()
```

Let's look at demand as a function of justice

```python
def distance_to_justice(t_1,t_2):
    temp=data.loc["R1",["D1","D2"]]*t_1+data.loc["R2",["D1","D2"]]*t_2
    temp/=temp.sum()
    return numpy.linalg.norm(temp-justice,ord=1)
```

```python
convextimes=[(t,1-t) for t in numpy.linspace(0,1)]
```

```python
ell=0.001
allowable=numpy.array([(s,t) for (s,t) in convextimes if distance_to_justice(s,t)<ell])
plt.figure()
plt.plot([0,1],[1,0],color="blue",linewidth=2)
plt.plot(allowable[:,0],allowable[:,1],color="red",label=r"$P_{{{0:.3f}}}$".format(ell),linewidth=5)
plt.xlabel("Demographic 1")
plt.ylabel("Demographic 2")
title=[]
title.append("Admissible allocations of time")
plt.title("\n".join(title))
legend("bottom")
saver("P001")
plt.show()
plt.close()

(d_infected,d_population)=(0.4,0.9)
a=analysis(d_infected,d_population)
hg=[a.compute_health_gap(t_1,t_2) for (t_1,t_2) in allowable]
hgmax=numpy.max(hg)
plt.plot(allowable[:,0],hg,color="blue",label="H(x,1-x)")
plt.plot(allowable[:,0],[0]*len(allowable),label="$P_{{{0:.3f}}}$".format(ell),color="red")
plt.axhline(hgmax,color="blue",linestyle="dotted",label=r"$\bar H({0:.1f},{1:.1f})={2:.2f}$".format(d_infected,d_population,hgmax))
plt.xlabel("Demographic 1 (x)")
plt.xlim(0,1)
title=[]
title.append("Health Gap")
title.append(r"$d_i={0:.1f}, d_p={1:.1f}$".format(d_infected,d_population))
plt.title("\n".join(title))
legend("bottom")
saver("healthgapa")
plt.show()
plt.close()
```
